```python
# 원하는 날짜의 강수량 데이터만 필터링
filtered_df = merged_df[merged_df['일시'].isin(desired_dates).reset_index(drop = True)]
filtered_df
```

```python
# 빈 데이터프레임 생성
result_df = pd.DataFrame(columns=['지점', '지점명', '일시', '강수량(mm)'])

# 원하는 날짜에 해당하는 시간별 강수량 확인
for date in desired_dates:
    date_df = merged_df[merged_df['일시'].dt.date == pd.to_datetime(date).date()]
    result_df = result_df.append(date_df[['지점', '지점명', '일시', '강수량(mm)']])

# 결과 출력
print(result_df)
```

```python
# 일자별 평균 강수량 계산
daily_avg = result_df.groupby(result_df['일시'].dt.date)['강수량(mm)'].mean()

# 일자별 최대 강수량 계산
daily_max = result_df.groupby(result_df['일시'].dt.date)['강수량(mm)'].max()
```
