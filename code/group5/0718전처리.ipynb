```python
data = pd.read_csv('100101_230713_rawdata.csv', encoding = 'cp949')
```

```python
data
```

```python
import datetime

data['일시']= pd.to_datetime(data['일시']) 
```

```python
converted_dates = pd.to_datetime(sorted_all_rain).date

# 필터링하여 새로운 DataFrame 생성
new_df = data[data['일시'].dt.date.isin(converted_dates)]

# 결과 출력
print(new_df)
```

```python
# 10년 1월 1일 ~ 23년 7월 13일
# 야구 경기할 때 비가 온 날들
# 해당 날에 1시간이라도 왔으면 다 포함
# 그 날의 00시 ~ 23시 강수량이 시간대별로 수집되어 있는 csv
new_df.to_csv('all_rain_raw_100101_230713.csv', encoding = 'cp949')
```

```python
# 일자별 평균 강수량 계산
all_rain_raw_100101_230713_daily_avg = new_df.groupby(new_df['일시'].dt.date)['강수량(mm)'].mean()

# 일자별 최대 강수량 계산
all_rain_raw_100101_230713_daily_max = new_df.groupby(new_df['일시'].dt.date)['강수량(mm)'].max()
```
